```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from pandas_datareader import data as web
import datetime as dt
start = dt.datetime(2019, 12, 8) # start date from when I want the data for the traditional asset and cryptocurrency
end = dt.datetime(2021, 12, 31)
# Importing the data
df = web.DataReader('BTC-USD', 'yahoo', start, end)
```

```python
df
```

```python
# Creating a variable for predicting 'n' days out into the future
projection = 14
# Create new column called prediction
df['Prediction'] = df[['Close']].shift(-projection)
```

```python
# Show the dataset
df
```

```python
# Create the independent data set (x)
X = np.array(df[['Close']])
# Remove the last 14 
X = X[:-projection]
print(X)
```

```python
# Create the dependent data set (y)
y = df['Prediction'].values
y = y[:-projection]
print(y)
```

```python
# Split the data inot 85% training and 15% testing data sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = .15)
```

```python
# Create and train the model
linReg = LinearRegression()
# Train the model
linReg.fit(x_train, y_train)
```

```python
# Test the model using score
linReg_confidence = linReg.score(x_test, y_test)
print('Linear regression Confidence:', linReg_confidence)
```

```python
# Create a variable called x_projection and set it equal to the last 14 rows of the data from the original data set
x_projection  = np.array(df[['Close']])[-projection:]
print(x_projection)
```

```python
# Print the linear regression models prediction for the next 14 days
linReg_prediction = linReg.predict(x_projection)
print(linReg_prediction)
```
